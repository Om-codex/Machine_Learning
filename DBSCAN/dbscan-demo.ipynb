```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[1,2],[2,2],[2,3],[8,7],[8,8],[25,80]])
```

```python
plt.scatter(X[:,0],X[:,1],color = 'black')
plt.show()
```

```python
db = DBSCAN(eps = 3,min_samples=2)
db.fit(X)
db.labels_
```

### > Here first three points are labeled as 0 so they are one cluster
### > The next two points are labeled as 1 so they form an another cluster
### > The noisy point is labeled as -1

```python
#increasing the min_samples
db = DBSCAN(eps = 3,min_samples=3)
db.fit(X)
db.labels_
```

```python
#increasing the epsilon value
db = DBSCAN(eps = 10,min_samples=2)
db.fit(X)
db.labels_
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
import numpy as np

X,_ = make_circles(n_samples=1000,factor =.5,noise = .03,random_state=4)

dbscan = DBSCAN(eps = 0.1,min_samples=5)
clusters = dbscan.fit_predict(X)

plt.scatter(X[:,0],X[:,1],c = clusters,cmap = 'viridis',marker = 'o')
plt.title('DBSCAN Clustering on the Concentric Circle dataset')
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.show()
```

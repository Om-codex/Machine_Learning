```python
import numpy as np
```

```python
from sklearn import datasets
iris = datasets.load_iris
```

```python
list(iris.keys())
```

```python
X = iris['data'][:, 3:]
# y = (iris['target'] == 2).astype(np.int)
```

```python
# y is 1 if the flower is Iris-virginica (target == 2), otherwise it's 0
y = (iris['target'] == 2).astype(int)
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
log_reg.fit(X,y)
```

```python
X_new = np.linspace(0,3,1000).reshape(-1,1)
```

```python
X_new
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

plt.plot(X[y==0], y[y==0], "bs")
plt.plot(X[y==1], y[y==1], "g^")

y_proba = log_reg.predict_proba(X_new)
decision_boundary = X_new[y_proba[:, 1] >= 0.5][0][0]

plt.plot(X_new,y_proba[:, 1], 'g-',label = 'Iris virginica')
plt.plot(X_new,y_proba[:,0],'b--',label = 'Not Iris virginica')

plt.plot([decision_boundary, decision_boundary], [-0.02, 1.02], "k:", linewidth=2)
plt.text(decision_boundary - 0.5, 0.15, "Decision boundary", fontsize=14, ha="center")

plt.arrow(decision_boundary, 0.92, 0.2, 0, head_width=0.05, head_length=0.1, fc='g', ec='g')
plt.arrow(decision_boundary, 0.08, -0.2, 0, head_width=0.05, head_length=0.1, fc='b', ec='b')

plt.xlabel("Petal width (cm)", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, 3, -0.02, 1.02]) # Set axis limits
plt.show()
```

# Key Takeaways from the Plot
### It Shows Probabilities, Not Just Predictions: The graph's primary purpose is to visualize the probability that a flower belongs to a certain class. The solid green line shows the rising probability of being an Iris virginica as petal width increases, while the dashed blue line shows the falling probability.

### The S-Shape is the Sigmoid Curve: The classic "S" shape of the curves is the sigmoid function at work. This is the core of logistic regression, mapping any input value into a probability between 0 and 1.

### The Intersection is the Decision Boundary: The point where the two curves cross, at a probability of 0.5 (around a petal width of 1.6 cm), is the decision boundary. This is the critical threshold the model uses to make a final classification. If a flower's petal width is to the right of this line, it's classified as Iris virginica; if it's to the left, it's not.

### The Curve's Height is the Model's Confidence: The y-axis directly represents the model's confidence. At the far left and right ends of the graph, the curves are flat and near 0 or 1, indicating the model is very certain of its prediction. In the middle, where the slope is steep, the model's confidence changes rapidly, and it is less certain about its classification.

```python
#importing the modules here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.linear_model
```

```python
# got a dataset loaded
df = pd.read_csv('linear_regression_dataset.csv')
df.head()
```

```python
# providing the x and y cols for training
X = np.c_[df['Hours_Studied']]
y = np.c_[df['Exam_Score']]
```

```python
# let's have a look on the data
plt.style.use('fivethirtyeight')
df.plot(kind = 'scatter',x = 'Hours_Studied',y = 'Exam_Score')
plt.title('Linear Regression Example')
plt.ylabel('Exam Scores')
plt.xlabel('Hours Studied')
```

```python
# creating a model
model = sklearn.linear_model.LinearRegression()
```

```python
# fit the data to the model
model.fit(X, y)
```

```python
# now make a prediction based on the training data
X_new = [ [5.4]]
print(model.predict(X_new))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
# Simulated dataset (curved relationship)
np.random.seed(42)
X = np.linspace(0, 10, 50).reshape(-1, 1)  # Study hours
y = 2 + 1.5*X - 0.1*X**2 + np.random.randn(50, 1)*2  # Exam scores (quadratic trend)
```

```python
# ---- Underfitting model (Linear Regression) ----
linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred_linear = linear_model.predict(X)
```

```python
# ---- Better model (Polynomial Regression) ----
poly_model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
poly_model.fit(X, y)
y_pred_poly = poly_model.predict(X)
```

```python
# ---- Visualization ----
plt.scatter(X, y, color="blue", label="Data (Exam Scores)")
plt.plot(X, y_pred_linear, color="red", label="Linear Model (Underfit)")
plt.plot(X, y_pred_poly, color="green", label="Polynomial Model (Better)")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.legend()
plt.title("Underfitting Example: Linear vs Polynomial Regression")
plt.show()
```

```python
import re
clean = re.compile('<.*?>')
re.sub(clean,'',df.iloc[0].review)
```

```python
# Function to clean html tags
def clean_html(text):
  clean = re.compile('<.*?>')
  return re.sub(clean,'',text)
```

```python
df['review'] = df['review'].apply(clean_html)
```

```python
# Converting everything to lowercase
def convert_lower(text):
  return text.lower()
```

```python
df['review'] = df['review'].apply(convert_lower)
```

```python
# Function to remove special characters

def remove_special(text):
  x= ''

  for i in text:
    if i.isalnum():
      x = x+i
    else:
      x = x + ' '
  return x
```

```python
remove_special('hii i am om%mishra studying in * bsc.*i*t second year+_()@')
```

```python
df['review'] = df['review'].apply(remove_special)
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

ps = PorterStemmer()
stops = set(stopwords.words('english'))

def preprocess(text):
    # Lowercase
    text = text.lower()
    # Remove non-alphabetic characters
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize into words
    words = word_tokenize(text)
    # Remove stopwords and short words
    words = [ps.stem(w) for w in words if w not in stops and len(w) > 1]
    # Join back into a string
    return ' '.join(words)

# Apply to your DataFrame
df['review'] = df['review'].apply(preprocess)
```

```python
# Function to remove stop words
stops = set(stopwords.words('english'))
def remove_stopwords(text):
    # Lowercase the text
    text = text.lower()
    # Remove punctuation and non-alphabetic characters
    text = re.sub(r'[^a-z\s]', '', text)
    # Split into tokens
    words = text.split()
    # Remove stopwords
    filtered_words = [word for word in words if word not in stops]
    # Join back into a string
    return ' '.join(filtered_words)
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

```python
df.head(10)
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

def stem_words(word_list):
    return [ps.stem(word) for word in word_list]
                        # or ' '.join(stemmed) if you want a string
```

```python
df['review'] = df['review'].apply(stem_words)
```

```python
# Join Back
def join_back(list_input):
    return ' '.join(list_input)
```

```python
df['review'] = df['review'].apply(join_back)
```

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples = 4, n_features=1, n_informative = 1, n_targets=1, noise = 80,random_state =13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

### Let's apply OLS method

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red')
```

### Lets apply the Gradient Descent assuming the slope is constant and m = 78.35 and let's assume the starting value for intercept is b = 0 

```python
y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
plt.plot(X,y_pred,color ='cyan',label = 'b = 0')
```

```python
# Now implementing the whole formula
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
# Let's take learning rate = 0.1
lr = 0.1

step_size = loss_slope * lr
step_size
```

```python
# Calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),label ='OLS',color='red')
plt.plot(X,y_pred1,color ='green',label = 'b = {}'.format(b))
plt.plot(X,y_pred,color ='cyan',label = 'b = 0')
plt.legend()
plt.show()
```

```python
# We can start the value of b from any starting value
b = - 100
m = 78.35
lr = 0.01

epochs = 10
for i in range(epochs):
  loss_slope = -2 * np.sum(y -m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m* X + b
  plt.plot(X,y_pred)

plt.scatter(X,y)
```
